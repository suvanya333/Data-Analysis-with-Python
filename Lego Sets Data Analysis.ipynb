```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\LENOVO\Downloads\lego-analysis-master\lego-analysis-master\datasets\lego_sets.csv')
theme = pd.read_csv(r'C:\Users\LENOVO\Downloads\lego-analysis-master\lego-analysis-master\datasets\parent_themes.csv')

df.head()
```

##### Merge two datasets

```python
df_merged=df.merge(theme,left_on='parent_theme',right_on='name')

df_merged.drop(columns='name_y',inplace=True)

df_merged.head()
```

#### drop rows where set_num is null 

```python
df_merged[df_merged['set_num'].isnull()].shape

df_merged.dropna(subset='set_num',inplace=True)
```

### 1. What percentage of all licenced sets ever released were Star Wars themed?

```python
licensed = df_merged[df_merged['is_licensed']]

star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
the_force = int(100* star_wars.shape[0]/licensed.shape[0])
the_force
```

### 2. In which year was Star Wars not the most popular licensed theme (in terms of sets relased that year)?

```python
licensed['count'] = 1

year_count = licensed.groupby(['year','parent_theme']).sum('count')['count'].reset_index()

max_count = year_count.sort_values('count',ascending = False).drop_duplicates(['year'])

max_count.sort_values('year',inplace=True)
max_count
```
